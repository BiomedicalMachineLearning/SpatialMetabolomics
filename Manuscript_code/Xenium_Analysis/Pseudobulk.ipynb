```r
for (i in 1:length(s_obj.l)){
    for (j in c('Cancer cells', 'Proliferation')){
        s_obj <- s_obj.l[[i]]
        s_obj@meta.data$de <- paste0(gsub(" ", "_",gsub("dhb-tissue region ", "", s_obj@meta.data$sample_SPM)), "_",s_obj@meta.data$cell_type)
        s_obj <- subset(x = s_obj, subset = cell_type == j)
        cell_metadata <- s_obj@meta.data
        filtered.sce <- SingleCellExperiment(assays = list(counts = s_obj@assays$SPM$counts),
            colData = cell_metadata)
        summed <- aggregateAcrossCells(filtered.sce, 
            id=colData(filtered.sce)[,c("orig.ident", "de","cell_type")])
        colnames(summed) <- colData(summed)$de
        counts.summed <- summed@assays@data$counts
        colnames(counts.summed) <- colData(summed)$de
        if (i==1 & j=='Cancer cells'){
        counts.summed.all <- counts.summed
        counts.summed.all.meta <-summed@colData
        }else{
        counts.summed.all <- cbind(counts.summed.all, counts.summed)
        counts.summed.all.meta <- rbind(counts.summed.all.meta, summed@colData)
        }
}
}
```

```r
for (i in 1:length(s_obj.l)){
    for (j in c('Oligodendrocyte','Astrocyte','Cancer cells','Microglia','Proliferation')){
        s_obj <- s_obj.l[[i]]
        s_obj@meta.data$de <- paste0(gsub(" ", "_",gsub("dhb-tissue region ", "", s_obj@meta.data$sample_SPM)), "_",s_obj@meta.data$cell_type)
        s_obj <- subset(x = s_obj, subset = cell_type == j)
        cell_metadata <- s_obj@meta.data
        filtered.sce <- SingleCellExperiment(assays = list(counts = s_obj@assays$SPM$counts),
            colData = cell_metadata)
        summed <- aggregateAcrossCells(filtered.sce, 
            id=colData(filtered.sce)[,c("orig.ident", "de","cell_type")])
        colnames(summed) <- colData(summed)$de
        counts.summed <- summed@assays@data$counts
        colnames(counts.summed) <- colData(summed)$de
        if (i==1 & j=='Oligodendrocyte'){
        counts.summed.all <- counts.summed
        counts.summed.all.meta <-summed@colData
        }else{
        counts.summed.all <- cbind(counts.summed.all, counts.summed)
        counts.summed.all.meta <- rbind(counts.summed.all.meta, summed@colData)
        }
}
}
```

```r
design <- model.matrix(~y$samples$groups)
design
```

```r
y <- estimateDisp(y, design)
summary(y$trended.dispersion)
```

```r
fit <- glmQLFit(y, design, robust=TRUE)
summary(fit$var.prior)
summary(fit$df.prior)
res <- glmQLFTest(fit, coef=ncol(design))
summary(decideTests(res))
```
