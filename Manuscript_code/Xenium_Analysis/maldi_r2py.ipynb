```r
library("Seurat")
library("arrow")
library(Matrix)
```

```r
files <- list.files(path="/scratch/project_mnt/S0010/Xiao/Q1851/Xiao/Working_project/MB/MALDI_PROCESSED", pattern="*.RDS", full.names=TRUE, recursive=FALSE)
```

```r
files_out_obsmeta <- sub(".RDS", "_obsmeta.parquet", files)
files_out_varmeta <- sub(".RDS", "_varmeta.parquet", files)
files_out_mtx <- sub(".RDS", ".mtx", files)
```

```r
for (i in 1:length(files)) {
  print(files[i])
  data <- readRDS(files[i])
  data@meta.data$xenium_id <- rownames(data@meta.data)
  arrow::write_parquet(data@meta.data, files_out_obsmeta[i])
  arrow::write_parquet(data@assays$SPM@meta.data, files_out_varmeta[i])
  Matrix::writeMM(data@assays$SPM@layers$counts, files_out_mtx[i])
}
```
